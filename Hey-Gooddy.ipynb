```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL to scrape
url = 'https://www.heygoody.com/checkinsurance/car'

# Request the content of the page
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Function to extract options from a select element
def extract_options(select_element):
    if select_element:
        return [option.text for option in select_element.find_all('option') if option.get('value')]
    return []

# Extract insurance companies
insurance_companies_select = soup.find('select', id='ddlbrand2')
insurance_companies = extract_options(insurance_companies_select)

# Placeholder functions to extract other details, replace 'select-id' with actual ids or classes
def extract_insurance_types(soup):
    insurance_types_select = soup.find('select', id='select-insurance-type')  # Replace with actual id or class
    return extract_options(insurance_types_select)

def extract_coverage(soup):
    coverage_select = soup.find('select', id='select-coverage')  # Replace with actual id or class
    return extract_options(coverage_select)

def extract_prices(soup):
    price_elements = soup.find_all('span', class_='price-class')  # Replace with actual class
    return [price.text for price in price_elements]

def extract_car_models(soup):
    car_models_select = soup.find('select', id='select-car-model')  # Replace with actual id or class
    return extract_options(car_models_select)

# Extract additional information
insurance_types = extract_insurance_types(soup)
coverage_details = extract_coverage(soup)
prices = extract_prices(soup)
car_models = extract_car_models(soup)

# Normalize lengths by finding the maximum length
max_length = max(len(insurance_companies), len(insurance_types), len(coverage_details), len(prices), len(car_models))

# Function to pad lists to maximum length
def pad_list(lst, length):
    return lst + [lst[-1] if lst else "Data not available"] * (length - len(lst))

# Pad all lists
insurance_companies = pad_list(insurance_companies, max_length)
insurance_types = pad_list(insurance_types, max_length)
coverage_details = pad_list(coverage_details, max_length)
prices = pad_list(prices, max_length)
car_models = pad_list(car_models, max_length)

# Create a DataFrame to store the data
data = {
    'Insurance Company': insurance_companies,
    'Insurance Type': insurance_types,
    'Coverage': coverage_details,
    'Price': prices,
    'Car Model': car_models
}

df = pd.DataFrame(data)

# Print the DataFrame
print(df)

# Optionally, save the DataFrame to a CSV file
df.to_csv('insurance_data.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize the WebDriver
driver = webdriver.Chrome()

# URL to scrape
url = 'https://www.heygoody.com/checkinsurance/car'
driver.get(url)

# Wait for the page to load
wait = WebDriverWait(driver, 20)

# Function to extract options from a select element
def extract_options(select_element):
    return [option.text for option in select_element.find_elements(By.TAG_NAME, 'option') if option.get_attribute('value')]

# Function to select an option by text
def select_option_by_text(select_element, text):
    for option in select_element.find_elements(By.TAG_NAME, 'option'):
        if option.text == text:
            option.click()
            break

# Function to handle the selection of an option
def safe_select_option_by_text(element_id, text):
    try:
        select_element = wait.until(EC.presence_of_element_located((By.ID, element_id)))
        select_option_by_text(select_element, text)
        return True
    except Exception as e:
        print(f"Error selecting {text} in element with ID {element_id}: {e}")
        return False

# Placeholder for results
results = []

try:
    # Extract and select car brands
    insurance_companies_select = wait.until(EC.presence_of_element_located((By.ID, 'ddlbrand2')))
    insurance_companies = extract_options(insurance_companies_select)

    for company in insurance_companies:
        if not safe_select_option_by_text('ddlbrand2', company):
            continue

        # Wait for car models to load and select a model
        car_models_select = wait.until(EC.presence_of_element_located((By.ID, 'ddlmodel')))
        car_models = extract_options(car_models_select)

        for model in car_models:
            if not safe_select_option_by_text('ddlmodel', model):
                continue

            # Wait for sub-models to load if applicable
            try:
                sub_models_select = wait.until(EC.presence_of_element_located((By.ID, 'ddlsubmodel')))
                sub_models = extract_options(sub_models_select)
                for sub_model in sub_models:
                    if not safe_select_option_by_text('ddlsubmodel', sub_model):
                        continue
            except:
                sub_model = "N/A"

            # Select an insurance company
            insurance_types_select = wait.until(EC.presence_of_element_located((By.ID, 'ddltype')))
            insurance_types = extract_options(insurance_types_select)

            for insurance_type in insurance_types:
                if not safe_select_option_by_text('ddltype', insurance_type):
                    continue

                # Select a start date
                try:
                    start_date_input = wait.until(EC.presence_of_element_located((By.ID, 'start_date')))
                    start_date_input.clear()
                    start_date_input.send_keys("01/01/2024")  # Example date
                except Exception as e:
                    print(f"Error entering start date: {e}")
                    continue

                # Select a province
                if not safe_select_option_by_text('ddlprovince', "Bangkok"):  # Example province
                    continue

                # Enter birth year
                try:
                    birth_year_input = wait.until(EC.presence_of_element_located((By.ID, 'birth_year')))
                    birth_year_input.clear()
                    birth_year_input.send_keys("1980")  # Example birth year
                except Exception as e:
                    print(f"Error entering birth year: {e}")
                    continue

                # Submit the form
                try:
                    submit_button = driver.find_element(By.ID, 'submit_button')
                    submit_button.click()
                except Exception as e:
                    print(f"Error clicking submit button: {e}")
                    continue

                # Wait for the results to load and extract details
                try:
                    time.sleep(5)  # Adjust sleep time as necessary
                    coverage_details = driver.find_element(By.CLASS_NAME, 'coverage-class').text  # Replace with actual class
                    price = driver.find_element(By.CLASS_NAME, 'price-class').text  # Replace with actual class

                    # Save the results
                    results.append({
                        'Insurance Company': company,
                        'Car Model': model,
                        'Sub Model': sub_model,
                        'Insurance Type': insurance_type,
                        'Coverage': coverage_details,
                        'Price': price
                    })

                except Exception as e:
                    print(f"Error extracting insurance details for {insurance_type} of model {model} of company {company}: {e}")

finally:
    # Convert results to a DataFrame
    df = pd.DataFrame(results)

    # Print the DataFrame
    print(df)

    # Optionally, save the DataFrame to a CSV file
    df.to_csv('insurance_data.csv', index=False)

    # Close the WebDriver
    driver.quit()
```
